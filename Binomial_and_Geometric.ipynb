```python
import numpy as np
import pandas as pd
import seaborn as sns

from scipy.stats import binom, geom
```

```python
p = 0.1 # success probability
x_vals = [0, 1] # no. of offer letters

probs = [1-p, p] # probability for 0 and 1 letter
probs
```

P(X=1) = p = 0.1 and
P(X=0) = 1-p = 0.9

### Probability Mass Function (PMF)

```python
sns.barplot(x=x_vals, y=probs)
```

## 2 Interviews (2 trials)

```python
x_vals = [0,1,2]
```

```python
probs = [
    (1-p) * (1-p),
    (1-p)*p + p*(1-p),
    p*p
]
probs
```

```python
sns.barplot(x=x_vals, y=probs)
```

## Three Trials

n = 3

```python
x_vals = [0,1,2,3]
n = 2
p = 0.1
```

```python
probs = [
    (1-p)*(1-p)*(1-p), #X=0
    (1-p)*(1-p)*p + (1-p)*p*(1-p) + p*(1-p)*(1-p), #X=1
    (1-p)*p*p + p*p*(1-p) + p*(1-p)*p, #X=2
    p*p*p, #X=3
]
```

```python
probs
```

```python
sns.barplot(x=x_vals, y=probs)
```

## binom.pmf()

```python
probs_binom = binom.pmf(x_vals, n=3, p=0.1)
probs_binom
```

```python
sns.barplot(x=x_vals, y=probs_binom)
```

## N trials

```python
n = 15
p = 0.5
x_vals = np.arange(0, n+1)
probs_binom = binom.pmf(x_vals, n=n, p=p)
sns.barplot(x=x_vals, y=probs_binom)
```

You have 10 interviews, what is the probability that you would succeed in at most 5 interviews, given that probability of success is 0.2?

```python
n = 10
k = 5
p = 0.2
X = [0,1,2,3,4,5]
probs = binom.pmf(k = X, n=n, p=p)
np.sum(probs)
```

## Geometric Distribution

```python
p = 0.1
x_vals = np.arange(1,20)
probs_geom = geom.pmf(x_vals, p=p)
probs_geom
```

```python
sns.barplot(x=x_vals, y=probs_geom)
```
